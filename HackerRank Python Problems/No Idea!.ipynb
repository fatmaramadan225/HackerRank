You are given:

- An array of **N integers**
- Two **disjoint sets**:
  - Set **A**: you like the elements in this set (+1 happiness)
  - Set **B**: you dislike the elements in this set (âˆ’1 happiness)

Your initial happiness is **0**.

---

## ðŸ’¡ Concept

- For each element in the array:
  - If the element is in set **A**, add **+1** to your happiness.
  - If the element is in set **B**, subtract **1** from your happiness.
  - If it's in neither, your happiness stays the same.
- Sets **A** and **B** contain no repeated elements.
- The array **can** contain duplicates.

---

## ðŸ“Œ Constraints

- All elements are integers.
- The values in A and B are **disjoint**.
- The array size and set sizes follow the problem definition.

---

## ðŸ§· Input Format

1. The first line contains **two integers**, `n` and `m`, separated by a space.
2. The second line contains **n space-separated integers** (the array).
3. The third line contains **m space-separated integers** (elements of set A).
4. The fourth line contains **m space-separated integers** (elements of set B).

---

## ðŸ“¤ Output Format

Output a single integer â€” your **final happiness score**.

## ðŸ§ª Sample Input

## âœ… Sample Output

## ðŸ§  Explanation

- Array: `[1, 5, 3]`
- Set A (like): `{3, 1}`
- Set B (dislike): `{5, 7}`

### Step-by-step:

- `1` â†’ in A â†’ happiness += 1
- `5` â†’ in B â†’ happiness -= 1
- `3` â†’ in A â†’ happiness += 1

### Final Happiness:  
`+1 -1 +1 = 1`

---

## ðŸ§° Tip

Use Python sets for A and B, and loop over the array:

```python
happiness = sum((i in A) - (i in B) for i in arr)
```

This one-liner uses Python's boolean arithmetic:
- `True` â†’ 1
- `False` â†’ 0

---


````python
n,m =list(map(int, input().split()))

arr = list(map(int, input().split()))
A = set(map(int, input().split()))
B = set(map(int, input().split()))
happiness =0
for i in arr:
    if i in A:
        happiness +=1
    if i in B:
        happiness -=1
            
print(happiness)
````
