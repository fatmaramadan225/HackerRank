### collections.Counter()

A `Counter` is a container from the `collections` module that stores elements as dictionary keys, and their counts as dictionary values.

---

#### Sample Code

```python
from collections import Counter

myList = [1, 1, 2, 3, 4, 5, 3, 2, 3, 4, 2, 1, 2, 3]

print(Counter(myList))
# Output: Counter({2: 4, 3: 4, 1: 3, 4: 2, 5: 1})

print(Counter(myList).items())
# Output: [(1, 3), (2, 4), (3, 4), (4, 2), (5, 1)]

print(Counter(myList).keys())
# Output: [1, 2, 3, 4, 5]

print(Counter(myList).values())
# Output: [3, 4, 4, 2, 1]
```

### Task

A shoe shop owner has `n` shoes in stock.  
Each shoe has a specific size.

There are `x` customers, each willing to pay a certain price **only if** the desired shoe size is available.

Your task is to compute how much money the shop owner earns by selling shoes to the customers.

---

### Input Format

- The first line contains `n`, the total number of shoes.
- The second line contains a space-separated list of the shoe sizes in the shop.
- The third line contains `x`, the number of customers.
- The next `x` lines contain two space-separated values:  
  the desired shoe size and the offered price.

---

### Constraints

- `0 < n < 10^3`  
- `0 < size <= 100`  
- `0 < x <= 10^3`  
- `0 < price <= 1000`

---

### Output Format

- Print the total amount of money earned.

---

### Sample Input
```
10
2 3 4 5 6 8 7 6 5 18
6
6 55
6 45
6 55
4 40
18 60
10 50
```

---

### Sample Output
```
200
```

---

### Explanation

- Customer 1: Bought size 6 for \$55 ✅  
- Customer 2: Bought size 6 for \$45 ✅  
- Customer 3: Size 6 out of stock ❌  
- Customer 4: Bought size 4 for \$40 ✅  
- Customer 5: Bought size 18 for \$60 ✅  
- Customer 6: Size 10 not available ❌

**Total earned: 55 + 45 + 40 + 60 = 200**


````python
from collections import Counter
num_shoes = int(input())
shoe_size = list(map(int, input().split()))
num_customer = int(input())


i = Counter(shoe_size)

earning = 0

for _ in range(num_customer):
    size, price = map(int, input().split())
    if i[size] > 0:
        earning += price
        i[size] -= 1 

print(earning)
````
