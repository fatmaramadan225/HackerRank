### String Validation

**Task**  
You are given a string `s`.  
Your task is to find out if the string `s` contains:  
- alphanumeric characters,  
- alphabetical characters,  
- digits,  
- lowercase and  
- uppercase characters.

---

**String methods**  
Python has built-in string validation methods for basic data types. These are:

- `str.isalnum()`: Returns True if all characters are alphanumeric (a-z, A-Z and 0-9).  
  ```python
  'ab123'.isalnum() → True  
  'ab123#'.isalnum() → False  
  ```

- `str.isalpha()`: Returns True if all characters are alphabetic (a-z and A-Z).  
  ```python
  'abcD'.isalpha() → True  
  'abcd1'.isalpha() → False  
  ```

- `str.isdigit()`: Returns True if all characters are digits (0-9).  
  ```python
  '1234'.isdigit() → True  
  '123edsd'.isdigit() → False  
  ```

- `str.islower()`: Returns True if all characters are lowercase (a-z).  
  ```python
  'abcd123#'.islower() → True  
  'Abcd123#'.islower() → False  
  ```

- `str.isupper()`: Returns True if all characters are uppercase (A-Z).  
  ```python
  'ABCD123#'.isupper() → True  
  'Abcd123#'.isupper() → False  
  ```

---

**Input Format**  
A single line containing a string `s`.

---

**Constraints**  
- The string contains only ASCII characters.

---

**Output Format**  
Print the following five lines:

1. Print `True` if `s` has any alphanumeric characters. Otherwise, print `False`.  
2. Print `True` if `s` has any alphabetical characters. Otherwise, print `False`.  
3. Print `True` if `s` has any digits. Otherwise, print `False`.  
4. Print `True` if `s` has any lowercase characters. Otherwise, print `False`.  
5. Print `True` if `s` has any uppercase characters. Otherwise, print `False`.

---

**Sample Input**  
```
qA2
```

**Sample Output**  
```
True  
True  
True  
True  
True
```


````python

s = input()
    
print(any(c.isalnum() for c in s))
print(any(c.isalpha() for c in s))
print(any(c.isdigit() for c in s))
print(any(c.islower() for c in s))
print(any(c.isupper() for c in s))
````
