### üìò Function Description

Complete the `rangoli` function below.

#### Function Signature
```python
def print_rangoli(size: int) -> None
```

#### Parameters
- `size` (int): the size of the rangoli (1 ‚â§ size ‚â§ 26)

#### Returns
- None (prints the rangoli pattern to standard output)

---

### üì• Input Format

A single line containing the integer `n`, the size of the rangoli.

---

### üì§ Output Format

Output the alphabet rangoli of the given size, centered and symmetrical.

---

### üîÅ Sample Input
```
5
```

### ‚úÖ Sample Output
```
--------e--------
------e-d-e------
----e-d-c-d-e----
--e-d-c-b-c-d-e--
e-d-c-b-a-b-c-d-e
--e-d-c-b-c-d-e--
----e-d-c-d-e----
------e-d-e------
--------e--------
```


````python
def print_rangoli(size):
    width = 4 * size - 3  
    for i in range(size-1):
        left = [chr(96 + size - j) for j in range(i + 1)]
        right = left[:-1][::-1]  
        row = '-'.join(left + right)
        print(row.center(width, '-'))
    for i in reversed(range(size)):
        left = [chr(96 + size - j) for j in range(i + 1)]
        right = left[:-1][::-1]  
        row = '-'.join(left + right)
        print(row.center(width, '-'))
        

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
````
