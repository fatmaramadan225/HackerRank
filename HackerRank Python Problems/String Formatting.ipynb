### ðŸ’¡ Tip

Use Python's built-in functions:
- `oct()`, `hex()`, `bin()` to get octal, hexadecimal, and binary representations.
- Use `.rjust(width)` to align all columns to the right based on the max binary length.



````python
def print_formatted(number):
    width = len(bin(number)[2:])
    for i in range(1, number + 1):
        dec_str = str(i).rjust(width)
        oct_str = oct(i)[2:].rjust(width)
        hex_str = hex(i)[2:].upper().rjust(width)
        bin_str = bin(i)[2:].rjust(width)
        print(f"{dec_str} {oct_str} {hex_str} {bin_str}")
````

````python
n = int(input())
print_formatted(n)
````
